```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('cheese_data.csv')
```

```python
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing numeric values with the mean
df['MoisturePercent'].fillna(df['MoisturePercent'].mean(), inplace=True)
```

```python
df.isnull().sum() # No more missing values in MoisturePercent
```

```python
df['ManufacturerProvCode'].value_counts()
```

```python
df['ManufacturingTypeEn'].value_counts()
```

```python
df['MilkTypeEn'].value_counts()
```

```python
# Filling the missing value of MilkTypeEn column
most_frequent_MilkTypeEn = df['MilkTypeEn'].mode()[0]
df['MilkTypeEn'].fillna(most_frequent_MilkTypeEn, inplace=True)
```

```python
df.isnull().sum()
```

```python
## Filling the missing value of CategoryTypeEn column
most_frequent_CategoryTypeEn = df['CategoryTypeEn'].mode()[0]
df['CategoryTypeEn'].fillna(most_frequent_CategoryTypeEn, inplace=True)
```

```python
df.isnull().sum()
```

```python
## Filling the missing value of MilkTreatmentTypeEn column
most_frequent_MilkTreatmentTypeEn = df['MilkTreatmentTypeEn'].mode()[0]
df['MilkTreatmentTypeEn'].fillna(most_frequent_MilkTreatmentTypeEn, inplace=True)
```

```python
df.isnull().sum()
```

```python
new_df=df[['ManufacturerProvCode','ManufacturingTypeEn','MoisturePercent','Organic','CategoryTypeEn','MilkTypeEn','MilkTreatmentTypeEn','CheeseName','FatLevel']]
```

```python
new_df.isnull().sum() # No missing value in this new dataset.
```

```python
new_df.duplicated().sum()
```

```python
new_df.drop_duplicates(inplace=True)
```

```python
mean_MoisturePercent = df['MoisturePercent'].mean()
median_MoisturePercent = df['MoisturePercent'].median()
mode_MoisturePercent = df['MoisturePercent'].mode().values
std_MoisturePercent = df['MoisturePercent'].std()
variance_MoisturePercent = df['MoisturePercent'].var()
min_MoisturePercent = df['MoisturePercent'].min()
max_MoisturePercent = df['MoisturePercent'].max()
quartiles = df['MoisturePercent'].quantile([0.25, 0.5, 0.75])
IQR = quartiles[0.75] - quartiles[0.25]

print(f'Mean MoisturePercent: {mean_MoisturePercent}')
print(f'Median MoisturePercent: {median_MoisturePercent}')
print(f'Mode MoisturePercent: {mode_MoisturePercent}')
print(f'Standard Deviation of MoisturePercent: {std_MoisturePercent}')
print(f'Variance of MoisturePercent: {variance_MoisturePercent}')
print(f'Minimum MoisturePercent: {min_MoisturePercent}, Maximum MoisturePercent: {max_MoisturePercent}')
print(f'Quartiles:\n{quartiles}')
print(f'Interquartile Range (IQR): {IQR}')
```
